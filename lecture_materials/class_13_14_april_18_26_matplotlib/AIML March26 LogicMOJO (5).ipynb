data = [
    (101, "Rahul", 80000, "India", "Fraud"),
    (102, "Amit", 30000, "India", "Safe"),
    (103, "John", 90000, "USA", "Fraud"),
    (101, "Rahul", 80000, "India", "Fraud"),
    (104, "Sara", 40000, "USA", "Safe"),
    (105, "Mike", 120000, "USA", "Fraud")
]

1. Remove duplicate records
2. Convert data into dictionary format
3. create a set of unique countries
4. identify high income customers (>80k)
5. count fraud customers
6. skip customers with income < 40k
7. stop processing if income > 1 lakh found
8. store final result as tuple

```python
data = [
    (101, "Rahul", 80000, "India", "Fraud"),
    (102, "Amit", 30000, "India", "Safe"),
    (103, "John", 90000, "USA", "Fraud"),
    (101, "Rahul", 80000, "India", "Fraud"),
    (104, "Sara", 40000, "USA", "Safe"),
    (105, "Mike", 120000, "USA", "Fraud")
]

unique_data = list(set(data))

customers = []
countries = set()
fraud_count = 0
results = []

for row in unique_data:
    id, name, income, country , status = row
    if income < 40000:
        continue
    if income > 100000:
        print("Very high value customer")
        break
    customer = {"id": id, "name": name, "country": country, "status": status}
    customers.append(customer)
    countries.add(country)
    if status == "Fraud":
        fraud_count += 1
    if income > 80000:
        if status == "Fraud":
            label = "High risk and high value"
        else:
            label = "High value safe"
    else:
        label = "Normal"
    output = (id, label)
    results.append(output)

print("Customers: ", customers)
print("Countries: ", countries)
print("Fraud Count: ", fraud_count)
print("Results: ", results)
```

```python
import numpy as np

age = np.array([25, 30, 35, 40])
salary = np.array([40000, 50000, 60000, 70000])
experience = np.array([2, 5, 7, 10])


age = age.reshape(-1, 1)         
salary = salary.reshape(-1, 1)   
experience = experience.reshape(-1, 1)


dataset = np.hstack((age, salary, experience))  
print("Final Dataset:\n", dataset)


salary_updated = salary * 1.10   # vectorized
print("\nUpdated Salary:\n", salary_updated)


max_salary = np.max(salary_updated)
salary_normalized = salary_updated / max_salary  
# broadcasting: each value divided by max
print("\nNormalized Salary:\n", salary_normalized)


print("\nAll Ages:", dataset[:, 0])
print("\nFirst Employee:", dataset[0])


print("\nMean Salary:", np.mean(salary))
print("Median Experience:", np.median(experience))  
print("Std Age:", np.std(age))  


print("\nIterating through dataset:")
for row in dataset: 
    print(row)
```

```python
plt.plot(df.index, df['total_bill'])
plt.title('Line plot of total bill')
plt.xlabel('Index (customer order)')
plt.ylabel('total bill')
plt.show()
```

```python
avg_bill_day = df.groupby('day')['total_bill'].mean()
plt.plot(avg_bill_day.index, avg_bill_day.values, color = 'green', linestyle = '--', marker = '*')
plt.title('Average total bill')
plt.xlabel('Day')
plt.ylabel('Average bill')
plt.show()
print(avg_bill_day)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(df['total_bill'], df['tip'], c = df['sex'].map({'Male': 0, 'Female': 1}), cmap = 'coolwarm', s = 80, alpha = 0.6, edgecolor = 'red', marker = 'o')
plt.title('Total bill vs tip', fontsize = 16, fontweight = 'bold')
plt.xlabel('Total bill', fontsize = 12)
plt.ylabel('Tip given', fontsize = 12)
plt.grid(True, linestyle = '--', alpha = 0.8)
plt.show()
```

```python
plt.figure(figsize=(4,3))

for gender, color in zip(['Male', 'Female'], ['blue', 'red']):
    subset = df[df['sex'] == gender]
    plt.scatter(subset['total_bill'], subset['tip'], label=gender, color=color)
plt.title('Total bill vs tip', fontsize = 16, fontweight = 'bold')
plt.xlabel('Total bill', fontsize = 12)
plt.ylabel('Tip given', fontsize = 12)
plt.legend()
plt.grid(True, linestyle = '--', alpha = 0.8)
plt.show()
```
