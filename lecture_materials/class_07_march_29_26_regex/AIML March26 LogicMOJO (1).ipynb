data = [
    (101, "Rahul", 80000, "India", "Fraud"),
    (102, "Amit", 30000, "India", "Safe"),
    (103, "John", 90000, "USA", "Fraud"),
    (101, "Rahul", 80000, "India", "Fraud"),
    (104, "Sara", 40000, "USA", "Safe"),
    (105, "Mike", 120000, "USA", "Fraud")
]

1. Remove duplicate records
2. Convert data into dictionary format
3. create a set of unique countries
4. identify high income customers (>80k)
5. count fraud customers
6. skip customers with income < 40k
7. stop processing if income > 1 lakh found
8. store final result as tuple

```python
data = [
    (101, "Rahul", 80000, "India", "Fraud"),
    (102, "Amit", 30000, "India", "Safe"),
    (103, "John", 90000, "USA", "Fraud"),
    (101, "Rahul", 80000, "India", "Fraud"),
    (104, "Sara", 40000, "USA", "Safe"),
    (105, "Mike", 120000, "USA", "Fraud")
]

unique_data = list(set(data))

customers = []
countries = set()
fraud_count = 0
results = []

for row in unique_data:
    id, name, income, country , status = row
    if income < 40000:
        continue
    if income > 100000:
        print("Very high value customer")
        break
    customer = {"id": id, "name": name, "country": country, "status": status}
    customers.append(customer)
    countries.add(country)
    if status == "Fraud":
        fraud_count += 1
    if income > 80000:
        if status == "Fraud":
            label = "High risk and high value"
        else:
            label = "High value safe"
    else:
        label = "Normal"
    output = (id, label)
    results.append(output)

print("Customers: ", customers)
print("Countries: ", countries)
print("Fraud Count: ", fraud_count)
print("Results: ", results)
```
